```python
import pandas as pd
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt
import seaborn as sns

def black_scholes_call(S, K, T, r, sigma):
    if T <= 0 or sigma <= 0:
        return max(S - K, 0)
    d1 = (np.log(S / K) + (r + 0.5*sigma**2) * T) / (sigma * np.sqrt(T))
    d2 = d1 - sigma * np.sqrt(T)
    call = S * norm.cdf(d1) - K * np.exp(-r*T) * norm.cdf(d2)
    return max(call, max(S - K, 0))

def black_scholes_put(S, K, T, r, sigma):
    if T <= 0 or sigma <= 0:
        return max(K - S, 0)
    d1 = (np.log(S / K) + (r + 0.5*sigma**2) * T) / (sigma * np.sqrt(T))
    d2 = d1 - sigma * np.sqrt(T)
    put = K * np.exp(-r*T) * norm.cdf(-d2) - S * norm.cdf(-d1)
    return max(put, max(K - S, 0))


def mse(y_true, y_pred):
    return np.mean((y_true - y_pred)**2)

def mae(y_true, y_pred):
    return np.mean(np.abs(y_true - y_pred))

def mape(y_true, y_pred, epsilon=0.01):
    y_true = np.array(y_true)
    y_pred = np.array(y_pred)
    return np.mean(np.abs((y_true - y_pred) / np.maximum(y_true, epsilon))) * 100

def log_error(y_true, y_pred):
    return np.mean(np.abs(np.log1p(y_pred) - np.log1p(y_true)))


file_path = 'test_gld_after_2016_vola_2.xlsx'
T_days = 5
T = T_days / 252
r = 0.02

strike_cols = ['Strike_1','Strike_2','Strike_3','Strike_4','Strike_5']
close_cols = ['Close_1','Close_2','Close_3','Close_4','Close_5']


vol_cols = [
   '3_Day_Percent_Volatility', '9_Day_Percent_Volatility', '14_Day_Percent_Volatility',
  '21_Day_Percent_Volatility', '30_Day_Percent_Volatility', '60_Day_Percent_Volatility',
   '90_Day_Percent_Volatility'
 
]




def process_sheet(sheet_name, option_type='call', vol_col='14_Day_Percent_Volatility'):
    df = pd.read_excel(file_path, sheet_name=sheet_name, parse_dates=['Exp Date = T'])
    metrics_list = []

    all_close_values = df[close_cols].values.flatten()
    MOP = np.mean(all_close_values)

    for strike, close in zip(strike_cols, close_cols):
        pred_col = f'Predicted_{close}'
        bls_strike_col = f'BLS_{strike}'
        df[bls_strike_col] = df[strike] 

        if option_type.lower() == 'call':
            df[pred_col] = df.apply(
                lambda row: black_scholes_call(
                    S=row['ETF Price on T - 5'],
                    K=row[strike],
                    T=T,
                    r=r,
                    sigma=row[vol_col]/100
                ), axis=1
            )
        else:
            df[pred_col] = df.apply(
                lambda row: black_scholes_put(
                    S=row['ETF Price on T - 5'],
                    K=row[strike],
                    T=T,
                    r=r,
                    sigma=row[vol_col]/100
                ), axis=1
            )

        mse_val = mse(df[close], df[pred_col])
        mae_val = mae(df[close], df[pred_col])
        mape_val = mape(df[close], df[pred_col])
        logerr_val = log_error(df[close], df[pred_col])
        MAE_MOP_pct_val = mae_val / MOP * 100

        metrics_list.append({
            'Close': close,
            'MSE': mse_val,
            'MAE': mae_val,
            'MAE_MOP_pct': MAE_MOP_pct_val,
            'MAPE': mape_val,
            'LogError': logerr_val
        })

    metrics_df = pd.DataFrame(metrics_list)
    Mean_metrics = metrics_df[['MSE','MAE','MAE_MOP_pct','MAPE','LogError']].mean()
    Mean_metrics_row = {'Close':'Mean'}
    Mean_metrics_row.update(Mean_metrics.to_dict())
    metrics_df = pd.concat([metrics_df, pd.DataFrame([Mean_metrics_row])], ignore_index=True)


    df['Mean_Actual'] = df[close_cols].mean(axis=1)
    pred_cols_list = [f'Predicted_{c}' for c in close_cols]
    df['Mean_Predicted'] = df[pred_cols_list].mean(axis=1)

    plt.figure(figsize=(14,6))
    plt.plot(df['Exp Date = T'], df['Mean_Actual'], label='Mean Actual', linestyle='-', marker='o')
    plt.plot(df['Exp Date = T'], df['Mean_Predicted'], label='Mean Predicted', linestyle='--', marker='x')
    plt.title(f'{sheet_name} - Mean Actual vs Predicted Option Prices ({vol_col})')
    plt.xlabel('Date')
    plt.ylabel('Mean Option Price')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    export_cols = ['Exp Date = T', 'T - 5', 'ETF Price on T - 5']
    for strike, close in zip(strike_cols, close_cols):
        export_cols.extend([strike, f'BLS_{strike}', close, f'Predicted_{close}'])
    df_export = df[export_cols]
    df_export['Vol_Used'] = vol_col

    return df_export, metrics_df


all_vol_metrics = [] 

for vol_column in vol_cols:
    print(f"\n=== Processing Volatility Column: {vol_column} ===")
    
    call_df, call_metrics_df = process_sheet('GLD_CALL', option_type='call', vol_col=vol_column)
    print("Call Metrics:")
    print(call_metrics_df)
    
    put_df, put_metrics_df = process_sheet('GLD_PUT', option_type='put', vol_col=vol_column)
    print("Put Metrics:")
    print(put_metrics_df)
    
    call_metrics_df['Vol_Used'] = vol_column
    put_metrics_df['Vol_Used'] = vol_column
    all_vol_metrics.append(pd.concat([call_metrics_df.assign(Type='Call'), put_metrics_df.assign(Type='Put')]))

    output_file = f"./BLS results/test/GLD_BLS_Predictions_{vol_column}.xlsx"
    with pd.ExcelWriter(output_file, engine='xlsxwriter') as writer:
        call_df.to_excel(writer, sheet_name='Call', index=False)
        call_metrics_df.to_excel(writer, sheet_name='Call Metrics', index=False)
        put_df.to_excel(writer, sheet_name='Put', index=False)
        put_metrics_df.to_excel(writer, sheet_name='Put Metrics', index=False)
    print(f"Saved results to '{output_file}'")


combined_metrics_df = pd.concat(all_vol_metrics, ignore_index=True)
Mean_metrics = combined_metrics_df.groupby('Type')[['MSE','MAE','MAE_MOP_pct','MAPE','LogError']].mean().reset_index()
print("\n=== Mean Metrics Across All Volatilities ===")
print(Mean_metrics)

Mean_metrics.to_excel('./BLS results/test/GLD_BLS_Mean_Metrics_All_Volatilities_test.xlsx', index=False)
print("Saved Mean metrics across all volatilities to 'GLD_BLS_Mean_Metrics_All_Volatilities_test.xlsx'")

def extract_days(vol_name):
    return int(vol_name.split('_')[0])

combined_metrics_df['Vol_Days'] = combined_metrics_df['Vol_Used'].apply(extract_days)
combined_metrics_df = combined_metrics_df.sort_values('Vol_Days')

metrics_to_plot = ['MAPE','MAE','MSE']


for metric in metrics_to_plot:
    plt.figure(figsize=(10,6))
    for option_type in ['Call','Put']:
        df_plot = combined_metrics_df[combined_metrics_df['Type'] == option_type]
        Mean_metric = df_plot.groupby('Vol_Days')[metric].mean().reset_index()
        plt.plot(Mean_metric['Vol_Days'], Mean_metric[metric], marker='o', label=f'{option_type} Mean {metric}')

    plt.title(f'Mean {metric} vs Volatility Window')
    plt.xlabel('Volatility Window (Days)')
    plt.ylabel(f'Mean {metric}')
    plt.xticks(Mean_metric['Vol_Days'])
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.show()


```

```python
import pandas as pd
import matplotlib.pyplot as plt
import re


mlp_mae = pd.read_csv('mlp/mlp_mae.csv', index_col=0)
mlp_mse = pd.read_csv('mlp/mlp_mse.csv', index_col=0)
mlp_mape = pd.read_csv('mlp/mlp_mape.csv', index_col=0)
mlp_mae_pct = pd.read_csv('mlp/mae_percent_table.csv', index_col=0)

mlp_dict = {
    'MAE': mlp_mae,
    'MSE': mlp_mse,
    'MAPE': mlp_mape,
    'MAE_MOP_pct': mlp_mae_pct
}

def get_vol_days(model_name):
    match = re.search(r'with (\d+)d v', model_name)
    return int(match.group(1)) if match else 0

def get_optimizer(model_name):
    return 'Adam' if 'Adam' in model_name else 'SGD'

for df in mlp_dict.values():
    df['Vol_Days'] = df.index.map(get_vol_days)
    df['Optimizer'] = df.index.map(get_optimizer)


def extract_days(vol_name):
    return int(vol_name.split('_')[0])

combined_metrics_df['Vol_Days'] = combined_metrics_df['Vol_Used'].apply(extract_days)
combined_metrics_df = combined_metrics_df.sort_values('Vol_Days')


metrics_to_plot = ['MAE','MSE','MAPE','MAE_MOP_pct']
fig, axes = plt.subplots(2, 2, figsize=(16,10))
axes = axes.flatten()
mid_term_range = (14, 30)
handles_labels = []


results_list = []


all_vols = sorted(combined_metrics_df['Vol_Days'].unique())

for i, metric in enumerate(metrics_to_plot):
    ax = axes[i]


    bls_call = combined_metrics_df[combined_metrics_df['Type']=='Call'].groupby('Vol_Days')[metric].mean().reset_index()
    bls_put  = combined_metrics_df[combined_metrics_df['Type']=='Put'].groupby('Vol_Days')[metric].mean().reset_index()

    bls_call_line, = ax.plot(bls_call['Vol_Days'], bls_call[metric], marker='o', linestyle='-', color='green')
    bls_put_line,  = ax.plot(bls_put['Vol_Days'], bls_put[metric], marker='o', linestyle='-', color='red')

    bls_call_mean = ax.axhline(bls_call[metric].mean(), linestyle='--', color='darkgreen')
    bls_put_mean  = ax.axhline(bls_put[metric].mean(), linestyle='--', color='darkred')

    
    mid_mask_call = (bls_call['Vol_Days'] >= mid_term_range[0]) & (bls_call['Vol_Days'] <= mid_term_range[1])
    mid_mask_put  = (bls_put['Vol_Days'] >= mid_term_range[0]) & (bls_put['Vol_Days'] <= mid_term_range[1])
    ax.scatter(bls_call['Vol_Days'][mid_mask_call], bls_call[metric][mid_mask_call], s=120, edgecolor='k', color='green', zorder=5)
    ax.scatter(bls_put['Vol_Days'][mid_mask_put],  bls_put[metric][mid_mask_put],  s=120, edgecolor='k', color='red', zorder=5)

    if i == 0:
        handles_labels.extend([
            (bls_call_line, 'BLS Call'),
            (bls_put_line, 'BLS Put'),
            (bls_call_mean, 'BLS Call Mean'),
            (bls_put_mean, 'BLS Put Mean')
        ])

    
    mlp_df = mlp_dict[metric]

    for opt in ['Adam','SGD']:
        mlp_opt = mlp_df[mlp_df['Optimizer']==opt]
        mlp_grouped = mlp_opt.groupby('Vol_Days')[['Call Test','Put Test']].mean().reset_index()

        color_call = 'purple' if opt=='Adam' else 'darkseagreen'
        color_put  = 'orange' if opt=='Adam' else 'salmon'

        mlp_call_line, = ax.plot(mlp_grouped['Vol_Days'], mlp_grouped['Call Test'], marker='x', linestyle='-', color=color_call)
        mlp_put_line,  = ax.plot(mlp_grouped['Vol_Days'], mlp_grouped['Put Test'], marker='x', linestyle='-', color=color_put)

        mid_mask_call = (mlp_grouped['Vol_Days'] >= mid_term_range[0]) & (mlp_grouped['Vol_Days'] <= mid_term_range[1])
        mid_mask_put  = (mlp_grouped['Vol_Days'] >= mid_term_range[0]) & (mlp_grouped['Vol_Days'] <= mid_term_range[1])
        ax.scatter(mlp_grouped['Vol_Days'][mid_mask_call], mlp_grouped['Call Test'][mid_mask_call], s=100, edgecolor='k', color=color_call, marker='x', zorder=5)
        ax.scatter(mlp_grouped['Vol_Days'][mid_mask_put],  mlp_grouped['Put Test'][mid_mask_put], s=100, edgecolor='k', color=color_put, marker='x', zorder=5)

        if i == 0:
            handles_labels.extend([
                (mlp_call_line, f'MLP_CALL_{opt}'),
                (mlp_put_line, f'MLP_PUT_{opt}')
            ])

  
    ax.set_xticks(all_vols)
    ax.set_xlim(min(all_vols)-1, max(all_vols)+1)
    ax.tick_params(axis='x', labelsize=14, rotation=45)
    ax.tick_params(axis='y', labelsize=14)

    ax.set_title(f'{metric} vs Volatility Window', fontsize=15)
    ax.set_xlabel('Volatility Window (Days)', fontsize=15)
    ax.set_ylabel(metric, fontsize=14)
    ax.grid(True)

   
    for vol in bls_call['Vol_Days']:
        row = {
            'Metric': metric,
            'Vol_Days': vol,
            'BLS_Call': bls_call[bls_call['Vol_Days']==vol][metric].values[0],
            'BLS_Put':  bls_put[bls_put['Vol_Days']==vol][metric].values[0],
        }
        for opt in ['Adam','SGD']:
            mlp_opt = mlp_df[mlp_df['Optimizer']==opt]
            mlp_grouped = mlp_opt.groupby('Vol_Days')[['Call Test','Put Test']].mean()
            row[f'MLP_Call_{opt}'] = mlp_grouped.at[vol, 'Call Test'] if vol in mlp_grouped.index else None
            row[f'MLP_Put_{opt}']  = mlp_grouped.at[vol, 'Put Test'] if vol in mlp_grouped.index else None
        results_list.append(row)


results_df = pd.DataFrame(results_list)
results_df.to_csv('./BLS results/test/BLS_vs_MLP_Test_Metrics_by_Volatility.csv', index=False)
print("Saved results to BLS_vs_MLP_Test_Metrics_by_Volatility.csv")


plt.suptitle('BLS vs MLP 8L 3600N (Adam & SGD)  Test (2016-2017)  Metrics by Volatility Window (GLD Options)', fontsize=18)
fig.text(0.5, 0.01, 'Note: GLD = SPDR Gold Shares ETF', ha='center', fontsize=12, style='italic')

handles, labels = zip(*handles_labels)
fig.legend(handles, labels, loc='upper right', bbox_to_anchor=(1.075, 0.6), fontsize=12)

plt.tight_layout(rect=[0,0.03,0.95,0.95])
plt.savefig('./BLS results/test/GLD_BLS_vs_MLP_Adam_SGD_Test_.pdf', bbox_inches='tight')
plt.show()



```

```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv("./BLS results/test/BLS_vs_MLP_Test_Metrics_by_Volatility.csv")

metrics = ["MAE", "MAPE"]

vol_windows = [0, 3, 9, 14, 21, 30, 60, 90]
vol_labels = ['Without Volatility Input', 3, 9, 14, 21, 30, 60, 90]

x_vals = list(range(len(vol_windows)))
vol_to_x = {v: i for i, v in enumerate(vol_windows)}


fig, axes = plt.subplots(2, 2, figsize=(16, 10))
axes = axes.flatten()

for i, metric in enumerate(metrics):

    
    ax = axes[i]
    sub = df[df["Metric"] == metric]

    ax.plot(sub["Vol_Days"], sub["BLS_Call"], marker="o", color="green", label="BLS Call")
    ax.plot(sub["Vol_Days"], sub["BLS_Put"], marker="o", color="red", label="BLS Put")

    ax.set_title(f"BlackScholes {metric} vs Volatility", fontsize=14, fontweight='bold')
    ax.set_xlabel("Volatility Days")
    ax.set_ylabel(metric)
    ax.set_xticks(vol_windows[1:])  
    ax.grid(True)
    ax.legend()


    ax = axes[i + 2]

    for opt, c_call, c_put in zip(
        ["Adam", "SGD"],
        ["purple", "darkseagreen"],
        ["orange", "salmon"]
    ):
        ax.plot(
            [vol_to_x[v] for v in sub["Vol_Days"]],
            sub[f"MLP_Call_{opt}"],
            marker="x",
            color=c_call,
            label=f"MLP Call {opt}"
        )
        ax.plot(
            [vol_to_x[v] for v in sub["Vol_Days"]],
            sub[f"MLP_Put_{opt}"],
            marker="x",
            color=c_put,
            label=f"MLP Put {opt}"
        )

    ax.set_title(f"MLP {metric} vs Volatility", fontsize=14, fontweight='bold')
    ax.set_xlabel("Volatility Days")
    ax.set_ylabel(metric)
    ax.set_xticks(x_vals)
    ax.set_xticklabels(vol_labels, rotation=45)
    ax.grid(True)
    ax.legend()



plt.suptitle(
    "Black-Scholes vs MLP\nMAE and MAPE Across Volatility Windows (Testing 2016-2017)",
    fontsize=15,
    fontweight="bold",
    y=0.99
)

plt.figtext(
    0.5, 0.92,
    "Dataset: SPDR Gold Shares (GLD) ETF Options, 2016-2017",
    ha='center',
    fontsize=11,
    style='italic'
)

plt.tight_layout(rect=[0, 0, 1, 0.93])
plt.savefig("./BLS results/test/BLS_vs_MLP_MAE_MAPE_testing.pdf", dpi=300)
plt.show()
```
